# Loading the pandas library

```
import pandas as pd
```

# Loading the Datasets from the link provided

##Loading the 1st dataset

```
#The first dataset from an excel file
df1 = pd.read_excel('cells_geo_description.xlsx')
df1
```

##Loading the 2nd dataset

```
#Loading the dataset as a csv file and splitting the columns using a delimeter
df2 = pd.read_csv('cells_geo.csv', delimiter=';', index_col=0)
df2.head(50)
```

##Loading the 3rd dataset

```
#The 3rd dataset is loaded as an excel file
df3 = pd.read_excel('CDR_description.xlsx')
df3
```

##Loading the 4th dataset

```
#This datset is loaded as a csv file
df4 = pd.read_csv('Telcom_dataset.csv')
df4
```

##Loading the 5th dataset

```
#This dataset is also loaded as a csv file
df5 = pd.read_csv('Telcom_dataset2.csv')
df5
```

##Loading the 6th dataset

```
#This also loaded as a csv file
df6 = pd.read_csv('Telcom_dataset3.csv')
df6
```

# Cleaning Data

```
#Renaming the PRODUTC and the DATETIME columns on telcom_dataset to PRODUCT and DATE_TIME respectively
df4.rename(columns = {'PRODUTC': 'PRODUCT' , 'DATETIME': 'DATE_TIME'}, inplace = True)
df4
```

```
#Renaming the columns 'DW_A_NUMBER' AND 'DW_B_NUMBER' on the telcom_datset2 dataframe
#to 'DW_A_NUMBER_INT' AND 'DW_B_NUMBER_INT' respectively
df5.rename(columns = {'DW_A_NUMBER' : 'DW_A_NUMBER_INT' , 'DW_B_NUMBER' : 'DW_B_NUMBER_INT'} , inplace=True)
df5
```

```
#Renaming the CELLID and the SIET_ID columns on the telcom_dataset3 dataframe to CELL_ID and SITE_ID respectively.
df6.rename(columns = {'CELLID' : 'CELL_ID' , 'SIET_ID' : 'SITE_ID'} , inplace=True)
df6
```

# Merging Datasets, Splitting, Deleting and Constructing new columns

```
#Merging the datasets from the three days i.e telcom_dataset, telcom_dataset2 and telcom_dataset3
df7 = pd.concat([df4, df5, df6] , ignore_index=True)
df7
```

```
#Splitting the DATE_TIME column into two to give us date and time as different columns
df7[['DATE','TIME']] = df7.DATE_TIME.str.split(expand=True)
df7
```

```
#Deleting the DATE_TIME column aftre creating two separate DATE and TIME columns
del df7['DATE_TIME']
df7.head(60)
```

```
#Merging the combined dataset(df7) and the cell_geo dataset
df8 = df7.merge(df2, how = 'right', on = 'CELL_ID')
df8
```

# The most used cities for the 3 days

```
#Calculating the most used cities
df2['VILLES'].value_counts().head(10)
```

# The most used city for the 3 days

```
#Finding the most used city
df2['VILLES'].value_counts().head(1)
```

# The most used cities during business and home hours

```
#Grouping the dataset to determine the number of times they were in service using the status column
df9 = df2.groupby(['VILLES'], as_index = False)["STATUS"].count()
#Sorting them in descending order in order to determin the most used cities
df9.sort_values(["STATUS","VILLES"], axis = 0, ascending=False, inplace=True)
df9.head(15)
```

# The total value of each product

```
#Finding the total value of each product i.e. data, voice and sms
df7.groupby(['PRODUCT'], as_index=False)["VALUE"].sum()
```

# Determining the number of cities in Abidjan or interior

```
#Finding the number of cities in Abidjan and Interior
df2.groupby(['LOCALISATION'], as_index=False)["VILLES"].count()
```
